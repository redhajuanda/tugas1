```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("dataset2.csv")
df = df.iloc[:,1:4]

def normalEquation(X,Y):
    xTx = X.T.dot(X)
    XtX = np.linalg.inv(xTx)
    XtX_xT = XtX.dot(X.T)
    theta = XtX_xT.dot(Y)
    return theta
```

7.Jika dibangun model regresiyi=f(xi1, xi2) =w0+w1xi1+w2xi2+w3(xi1xi2)+w4(xi1+xi2)3 makadihasilkan prakiraan weight vectorˆw= (w0, w1, w2, w3, w4) = (4.514,−1.678,1.842,−0.002,−0.005)

```python
# x = pd.DataFrame(df['x1'])
x = pd.DataFrame(df['x1'])
x.insert(0,'w0',1)
x['x2'] = df['x2']
x['x1*x2'] = df['x1']*df['x2']
x['(x1+x2)^3'] = pow(df['x1']+df['x2'],3)
X = x.as_matrix()
Y = df['y'].as_matrix()

normalEquation(X,Y)
```
